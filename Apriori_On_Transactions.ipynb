```python
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
data = pd.read_csv('../input/new_Reduced_sales.csv')
product_df=pd.read_csv('../input/products.csv',sep=';')
# data.info()
data=data.iloc[:,1:]
data=data.drop(columns=['SalesID','SalesDate'])
product_id_data=data['ProductID']
```

```python
product_list=product_df['ProductName'].tolist()
product_list_id=product_df['ProductID'].tolist()
print(product_list,product_list_id)
```

```python
data.head()
```

```python
product_id_data.replace(product_list_id,product_list,inplace=True)
product_id_data
```

```python
data['ProductID']=product_id_data

basket = (data
          .groupby(['CustomerID', 'ProductID'])['Quantity']
          .sum().unstack().reset_index().fillna(0)
          .set_index('CustomerID'))
```

```python
basket.head()
```

```python
def encode_units(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1
basket = basket.applymap(encode_units)
# basket_sets.drop('POSTAGE', inplace=True, axis=1)
basket
```

```python
frequent_itemsets = apriori(basket, min_support=0.4, use_colnames=True)
print(frequent_itemsets)
```

```python
# frequent_itemsets = apriori(basket, min_support=0.07, use_colnames=True)
# print(frequent_itemsets)kshitij

rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules
```
