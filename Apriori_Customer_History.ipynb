```python
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
data = pd.read_csv('./User_Product_Matrix_Test.csv')
product_df=pd.read_csv('./products.csv',sep=';')
product_df.head()
```

```python
data=data.iloc[:,2:]
data.columns=product_df['ProductName'].tolist()
data.head()
```

```python
def encode_units(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1
data = data.applymap(encode_units)
# basket_sets.drop('POSTAGE', inplace=True, axis=1)
data
```

```python
frequent_itemsets = apriori(data, min_support=0.07, use_colnames=True)
# print(frequent_itemsets)
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules
```

```python
rules[ (rules['lift'] >= 2) &
      (rules['confidence'] >= 0.5) ]
```
